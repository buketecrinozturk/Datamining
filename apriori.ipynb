Veri setini Drive'dan çekiyoruz

Verileri liste haline getiriyoruz

```python
transactions = []
for i in range(0, 7500):
 transactions.append([str(dataset.values[i,j]) for j in range(0,20)])
```

Apriori algoritması için kuralları belirliyoruz

```python
!pip install apyori 
from apyori import apriori
rules = apriori(transactions = transactions, min_support = 0.003, min_cinfidence = 0.2, min_lift = 3, min_length = 2, max_length = 2)
```

Kuralları listeleyip results değişkenine atıyoruz

```python
results = list(rules)
```

```python
results
```

Sonuçların görselleştirilmesi

```python
def inspect(results):
    lhs         =[tuple(result[2][0][0])[0] for result in results]
    rhs         =[tuple(result[2][0][1])[0] for result in results]
    supports    =[result[1] for result in results]
    confidences =[result[2][0][2] for result in results]
    lifts        =[result[2][0][3] for result in results]
    return list (zip(lhs, rhs, supports, confidences, lifts))
resultsinDataFrame = pd.DataFrame(inspect(results), columns = ["Left hand side", "Right hand side", "Support", "Confidence", "Lift"])
```

Son olarak, görselleştirmelerini kolaylaştırmak için bu değişkenleri tek bir veri çerçevesinde saklarız.

```python
resultsinDataFrame
```

**Şimdi, bu nihai çıktıları azalan lift sırasına göre sıralıyoruz.**

```python
resultsinDataFrame.nlargest(n = 12, columns = "Lift")
```
